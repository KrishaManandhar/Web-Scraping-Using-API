```python

# Make the HTTP request and parse the JSON response
r = requests.get('https://companieslogo.com/api/1.0/?getAll&api_key=31f8603b62f86d70b48089')
data = r.json()

# Extract the links, ignoring any images that don't have the expected structure
links = [image['svg']['big']['for_bright_background']['orig'] for image in data if 'svg' in image and 'big' in image['svg'] and 'for_bright_background' in image['svg']['big']]

# Print the links
print(links)
```

```python
for link in links:
  print(link)
```

```python
import os
os.makedirs('company_logos',exist_ok=True)
```

```python
for i, link in enumerate(links):
    new_link =  'https://companieslogo.com/' + link
    response = requests.get(new_link)
    filename = f'company_logos/logo_{i}.svg'
    with open(filename, 'wb') as f:
        f.write(response.content)
        print(f'Saved logo for company {i} to {filename}')
```
